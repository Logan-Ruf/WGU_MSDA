```python
import pandas as pd
# Put CSV data into Pandas Dataframe
filename = 'sales_records.csv'

# Read CSV file into Pandas Dataframe
df = pd.read_csv(filename)
df
```

```python
# Check if there are any duplicate Order ID's
duplicates = df.duplicated(subset='Order ID').sum()
print(f'There are {duplicates} duplicate Order ID\'s in the dataset.')
```
